```python
a=pl.CUBE("H.cube")
b=pl.CUBE("1order.txt")
c=pl.CUBE("whole_cube_file_delta_0.1_approx_zero_0.01.txt")
#d=pl.CUBE("whole_cube_file_delta_0.1_approx_zero_0.001.txt") # wrong one
e=pl.CUBE("whole_cube_file_delta_0.03_approx_zero_0.01.txt")
f=pl.CUBE("whole_cube_file_delta_0.03_approx_zero_0.001.txt")
g=pl.CUBE("whole_cube_file_delta_0.05_approx_zero_0.01.txt")
h=pl.CUBE("whole_cube_file_delta_0.05_approx_zero_0.001.txt")
```
