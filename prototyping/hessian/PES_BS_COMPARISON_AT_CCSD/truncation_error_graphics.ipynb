```python
dt_qzvpp=pd.read_pickle('data_hfmp2_def2_qzvpp')
dt_tzvp=pd.read_pickle('data_hfmp2_def2_tzvp')
dt_sto3g=pd.read_pickle("data_hfmp2_sto3G")
dt_sto6g=pd.read_pickle("data_hfmp2_sto6G")
dt_321g=pd.read_pickle("data_hfmp2_321G")
dt_631g=pd.read_pickle("data_hfmp2_631G")
dt_aug=pd.read_pickle("data_hfmp2_aug")
dt_ccpvtz=pd.read_pickle("data_hfmp2_ccpvtz")
```

```python
def plot(ax,dfp,title):
    dfp['err_alch']=np.abs(dfp['err_alch'])
    dfp['delta_rho_int']=np.asarray((dfp.delta_rho),dtype=int)
    dfp=dfp.sort_values(by='delta_rho')
    for name, group in dfp.groupby('alchemy'):
        x_pl=[]
        y_pl=[]
        for name2,group2 in group.groupby('delta_rho_int'):
            x_pl.append(name2)
            y_pl.append(group2.err_alch.median())
        ax.plot(x_pl,y_pl,label='APDFT'+str(name))
        #ax.scatter(group.delta_rho,group.err_alch, color='C%d' % int(name), s=2) 
    ax.set_yscale('log')
    ax.set_ylim(3*1e-4,100)
    xmaxhf={"STO-3G":.45,"STO-6G":.45,"3-21G":.45,"6-31G*":.45,"cc-pVTZ":.7,"aug-cc-pVQZ":.65,"def2-TZVP":.7,"def2-QZVPP":.65}
    xmaxmp2={"STO-3G":.35,"STO-6G":.35,"3-21G":.35,"6-31G*":.35,"cc-pVTZ":.35,"aug-cc-pVQZ":.35,"def2-TZVP":.35,"def2-QZVPP":.35}
    xmaxca={"STO-3G":.25,"STO-6G":.25,"3-21G":.25,"6-31G*":.25,"cc-pVTZ":.25,"aug-cc-pVQZ":.25,"def2-TZVP":.25,"def2-QZVPP":.25}

    ax.axhline(np.abs(dfp.e_HF-dfp.e_ccsd).mean(),ls='-',xmax=xmaxhf[title],label='HF MAE',color='C7',lw=2.)
    ax.axhline(np.abs(dfp.e_mp2-dfp.e_ccsd).mean(),ls='-.',xmax=xmaxmp2[title],label='MP2 MAE',color='C8',lw=2.5)
    ax.axhline(1/630,ls=':',xmax=xmaxca[title],label="1 kcal/mol",color='black',lw=2.5)
    
    
    #ax.set_yticklabels(np.logspace(-3,1,5) ,minor=False,fontsize=20)
    #ax.tick_params(labelsize=15)
    ax.text(.03,.975,title,va='top',horizontalalignment='left', transform=ax.transAxes,fontsize=15,bbox=dict(fc='white',alpha=1))
```
